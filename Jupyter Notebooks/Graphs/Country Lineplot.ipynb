```python
import numpy as np
import pandas as pd
```

```python
import requests
import pprint
```

```python
c_data1 = requests.get('https://api.covid19india.org/data.json')
```

```python
time_series1=[]
for i in range(len(c_data2['cases_time_series'])):
    time_series1.append(c_data2['cases_time_series'][i]['totalconfirmed'])
```

```python
time_series1
```

```python
time_series1 = np.array(time_series1)
```

```python
time_series1
```

```python
time_series1.shape
```

```python
# Same for the date
```

```python
time_series_date1=[]
for i in range(len(c_data2['cases_time_series'])):
    time_series_date1.append(c_data2['cases_time_series'][i]['date'])
```

```python
time_series_date1 = np.array(time_series_date1)
```

```python
time_series_date1
```

```python
time_series_date1.shape
```

```python
time_series_date2=[]

for i in time_series_date1:
    time_series_date2.append(i + "2020")
```

```python
time_series_date2 = np.array(time_series_date2)
```

```python
time_series_date2
```

```python
time_series_date3 = pd.to_datetime(time_series_date2, infer_datetime_format=True)
```

```python
time_series_date3 = np.array(time_series_date3)
```

```python
time_series_date3
```

```python
time_series_date3.shape
```

```python
dq1 = pd.DataFrame(data={'confirmed_cases':time_series1, 'dates':time_series_date3})
```

```python
dq1
```

```python
#Plotting the Timeseries graph of Spread of virus around the Country
```

```python
import plotly.graph_objects as go


fig = go.Figure(data=go.Scatter(x=dq1['dates'], y=dq1['confirmed_cases']))

fig.update_layout(template="plotly_dark", title_text='Timeseries showing the spread of corona virus in India')



fig.show()
fig.write_html("India_timeseries.html")
```
